```python
x=df.drop("Happiness Score",axis=1)
y=df["Happiness Score"]
```

```python
from  sklearn.preprocessing import power_transform
x_new=power_transform(x)
x_new=pd.DataFrame(x_new)
x_new.columns=["Economy (GDP per Capita)","Family","Health (Life Expectancy) ","Freedom","Trust (Government Corruption)","Generosity","Dystopia Residual"]
x_new.head()
```

```python
x_new.skew()
```

```python
y.head()
```

```python
x_new.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
x_scaler=scaler.fit_transform(x_new)
x_scaler
print(x_scaler.shape)
print(y.shape)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
max_r_score=0
final_r_state=0
for i in range(1,200):
    x_train, x_test, y_train, y_test = train_test_split(x_scaler, y,test_size=0.30,random_state = i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred =lr.predict(x_test)
    r2_scr=r2_score(y_pred,y_test)
    print("r2 score corresponding to ",i," is ",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=i
print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
```

```python
##so using linear regression for random state 126
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y,test_size=0.20,random_state =178)

def eval(model):
    model.fit(x_train,y_train)
    y_pred =model.predict(x_test)
    r2_scr=r2_score(y_pred,y_test)
    print(r2_scr)
    print(mean_squared_error(y_pred,y_test))
    print(mean_absolute_error(y_pred,y_test))
```

```python
##Prediction for Linear Regression with random_state as 126
lr=LinearRegression()
eval(lr)
```

```python
##We will use DecisionTreeRegressor with random_state as 126
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
eval(rfr)
```

```python
## we will also use SupportVectorRegressor with random_state as 126
from sklearn.svm import SVR
svr=SVR()
eval(svr)
```

```python
## we will use cross validation to avoid chance of overfitting
##we apply cross validation for linear regression
from sklearn.model_selection import cross_val_score
lor=LinearRegression()
val_score=cross_val_score(lor,x_scaler,y,cv=4,scoring="r2")
print(val_score.mean())
```

```python
#Checking the cross validation score for RandomForest Regressor
rfr=RandomForestRegressor()
val_score=cross_val_score(rfr,x_scaler,y,cv=4,scoring="r2")
print(val_score.mean())
```

```python
##Checking the cross validation score for Supportvector Regressor
svr=SVR()
val_score=cross_val_score(svr,x_scaler,y,cv=4,scoring="r2")
print(val_score.mean())
```

```python
##from the above croos_val_score of we can infer that Linear Regression the best model to proceed further
##We will try to tune parameters
```

```python
from sklearn.model_selection import GridSearchCV
parameter={"fit_intercept":[True,False],"n_jobs":[1,None,1],"positive":[True,False]}
```

```python
grid=GridSearchCV(estimator=lr,param_grid=parameter,cv=4)
grid.fit(x_scaler,y)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
#now lets fit the model using the best parameters
x_train, x_test, y_train, y_test = train_test_split(x_scaler, y,test_size=0.20,random_state =126)

lr=LinearRegression(fit_intercept=True,n_jobs=1,positive=True)

lr.fit(x_train,y_train)

print(lr.score(x_train,y_train))

pred=lr.predict(x_test)

r2_scr=r2_score(pred,y_test)

print(lr.score(x_test,y_test))

print("r2_score",r2_scr)

print("mean_squared_error",mean_squared_error(pred,y_test))

print("mean_absolute_error",mean_absolute_error(pred,y_test))
```

```python
import joblib

joblib.dump(lr,"happiness_score_pred.obj")
```
